```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_spotify=pd.read_csv("D:\SpotifyFeatures.csv")
df_spotify.head()
```

```python
df_spotify.info()
```

```python
pd.isnull(df_spotify).sum()
#checking the data for any null values
```

```python
sorted_df=df_spotify.sort_values('popularity',ascending=True).head(10)
sorted_df
#sorting the vlaues in ascending order in terms of their popularity .
```

```python
most_popular=df_spotify.query('popularity>80', inplace=False).sort_values('popularity', ascending=False).head(20)
most_popular
#looking for the the tracks which are apopular and there popularity is greater than 100
```

```python
least_popular=df_spotify.query('popularity<10', inplace=False).sort_values('popularity', ascending=True)
least_popular[:20]
#finding the top 20 least popular songs on spotify
```

```python
most_danceable_and_popular=df_spotify[(df_spotify.danceability>0.5)&(df_spotify.popularity>90)].sort_values('danceability',ascending=False).head(20)
most_danceable_and_popular[:20]
#looking for the songs which are dancable and also popular
```

```python
df_spotify['duration_s']=df_spotify['duration_ms'].apply(lambda x: round(x/1000))
df_spotify.drop('duration_ms', inplace=True,axis=1)
#converting the duration in seconds
```

```python
df_spotify.duration_s.head()
```

```python
cor_df = df_spotify.corr(method="pearson",numeric_only=True)
plt.figure(figsize=(18, 16))
heatmap = sns.heatmap(cor_df, annot=True, fmt=".2f", vmin=-1, vmax=1, center=0, cmap='Spectral', linewidths=1, linecolor='black')
plt.title('Correlation between different variables of a song')
```

```python
trial_df=df_spotify.sample(int(0.01*len(df_spotify)))
#taking out a sample of data from this large data frame
```

```python
print(len(trial_df))
```

```python
plt.figure(figsize=(10,10))
sns.regplot(data=trial_df, y='loudness',x='danceability',color="c").set(title='loudness vs danceability')
```

```python
plt.figure(figsize=(10,10))
sns.regplot(data=trial_df, y='loudness',x='energy',color="c").set(title='loudness vs energy')
```

```python
plt.figure(figsize=(10,10))
sns.regplot(data=trial_df, y='acousticness',x='energy',color="c").set(title='acousticness vs energy')
```

```python
plt.figure(figsize=(10,10))
sns.regplot(data=trial_df, y='acousticness',x='popularity',color="b").set(title='acousticness vs popularity')
```
